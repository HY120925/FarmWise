```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/processed/final_training_dataset.csv")

print("âœ… Dataset loaded. Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nInfo:")
print(df.info())
print("\nMissing Values:")
print(df.isnull().sum())

print("\nSummary statistics:")
print(df.describe(include="all"))

plt.figure(figsize=(12,6))
sns.countplot(y=df['crop'], order=df['crop'].value_counts().index)
plt.title("Crop Distribution")
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(x=df['soil_type'])
plt.title("Soil Type Distribution")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(df['yield_tons'], bins=20, kde=True)
plt.title("Yield Distribution (tons)")
plt.show()

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="YlGnBu")
plt.title("Correlation Heatmap")
plt.show()

plt.figure(figsize=(12,6))
sns.boxplot(x="region", y="yield_tons", data=df)
plt.title("Yield by Region")
plt.xticks(rotation=45)
plt.show()
```
